```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sys 
import torch
```

```python
strokes = numpy.load('data/strokes-py3.npy', allow_pickle=True)
```

```python
print("Shape:", strokes[0].shape)
print("Shape:", strokes[0][0].shape)
print("Shape:", strokes[0][2].shape)
print("Data type:", strokes[0].dtype)
print("Length:", len(strokes))
```

```python
data = strokes
print(f"Data type of the numpy array: {data.dtype}")

# 2. Inspect Unique Data Types
unique_data_types = {type(item) for item in data}
print(f"Unique data types within the numpy array: {unique_data_types}")

# 3. Inspect Sequence Lengths
sequence_lengths = [len(item) if isinstance(item, (list, np.ndarray)) else 0 for item in data]
plt.hist(sequence_lengths, bins=50)
plt.xlabel('Sequence Length')
plt.ylabel('Number of Sequences')
plt.title('Distribution of Sequence Lengths')
plt.show()

# 4. Check for NaNs or Infs
contains_nan = any([np.isnan(item).any() for item in data if isinstance(item, (list, np.ndarray))])
contains_inf = any([np.isinf(item).any() for item in data if isinstance(item, (list, np.ndarray))])
print(f"Contains NaN values: {contains_nan}")
print(f"Contains Inf values: {contains_inf}")

# 5. Inspect Unique Values (only for a sample for demonstration)
sample_sequence = data[0] if len(data) > 0 else []
print(f"Unique values in the first sequence: {np.unique(sample_sequence)}")
```

```python
# 1. Check for Non-Uniform Datatypes
datatypes = [item.dtype for item in data]
unique_datatypes = set(datatypes)

if len(unique_datatypes) > 1:
    print(f"Found multiple datatypes in the numpy array: {unique_datatypes}")
else:
    print(f"All entries have the same datatype: {list(unique_datatypes)[0]}")

# 2. Check for Non-Uniform Shapes
shapes = [item.shape for item in data]
unique_shapes = set(shapes)

if len(unique_shapes) > 1:
    print(f"Found multiple shapes in the numpy array. Here are some of them:")
    print(list(unique_shapes)[:10])  # print the first 10 unique shapes as an example
else:
    print(f"All entries have the same shape: {list(unique_shapes)[0]}")

# 3. Convert to a List of Tensors
list_of_tensors = [torch.tensor(item) for item in data]
```

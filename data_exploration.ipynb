```python

import math

import torch.optim as optim

def find_learning_rate(model, data_loader, criterion, init_value=1e-7, final_value=1e-1, beta=0.98, num_iter=100):

    # num = len(data_loader) - 1
    # print(num)
    mult = (final_value / init_value) ** (1/num_iter)
    lr = init_value
    # optimizer = torch.optim.SGD(model.parameters(), lr=lr)
    alpha = 0.95  # Equivalent to ξ (xi) in the paper
    momentum = 0.9  # Equivalent to ζ (zeta) in the paper
    eps = 0.0001  # Equivalent to ι (iota) in the paper
    optimizer = optim.RMSprop(
        model.parameters(), lr=lr, alpha=alpha, eps=eps, momentum=momentum)
    avg_loss = 0.
    best_loss = 0.
    losses = []
    log_lrs = []
    for batch_num, (xt, _) in enumerate(data_loader):
        # if xt.nelement() == 0:  # Check if the data tensor is empty
        #     print(f"No data in batch {batch_num}")
        #     break
        # if batch_num >= num_iter:  # Only iterate num_iter times
        #     break
        optimizer.param_groups[0]['lr'] = lr
        batch_num += 1 # prevent zero division errors
        outputs = model(xt[:,:-1,:])
        loss = criterion(outputs, xt[:,1:,:])

        # Compute the smoothed loss
        avg_loss = beta * avg_loss + (1-beta) * loss.item()
        smoothed_loss = avg_loss / (1 - beta**batch_num)
        
        # Stop if the loss is exploding
        if batch_num > 1 and smoothed_loss > 4 * best_loss:
            return log_lrs, losses
        
        if smoothed_loss < best_loss or batch_num==1:
            best_loss = smoothed_loss

        losses.append(smoothed_loss)
        log_lrs.append(lr)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        lr *= mult
    # print(f"Returning values from {len(log_lrs)} iterations")
    return log_lrs, losses
```

```python
from itertools import islice
import numpy as np
train_dataset = Stroke_dataset(train=True)
model = Uncondition_LSTM(
    alphabet_size=78,
    input_size=3,
    hidden_size=400,
    num_layers=3,
    component_K=20,
    dropout=0.1
).to('cpu')
model._initialize_weights()
full_data_loader = DataLoader(
        train_dataset, batch_size=1, shuffle=True, num_workers=4)
# print(len(data_loader))
data_loader = islice(full_data_loader, 100)
loss_fn = MDN_loss_function
log_lrs, losses = find_learning_rate(model, data_loader, loss_fn)


```

```python
import matplotlib.pyplot as plt
plt.plot(log_lrs, losses)
plt.xscale('log')  
plt.xlabel('Learning Rate')
plt.ylabel('Loss')
plt.title('Learning Rate Finder')
plt.show()

```

```python
from data.stroke_dataset import Stroke_dataset
from utils import plot_stroke_with_dataset
import matplotlib.pyplot as plt
import numpy as np

dataset = Stroke_dataset(train=True)

def plot_stroke_with_sentence(dataset, index, save_name=None):
    # Retrieve stroke and corresponding sentence from the dataset
    stroke, encoded_sentence = dataset[index]
    stroke = stroke.numpy()  # Convert to numpy array if it's a tensor

    # Decoding the sentence (optional, if you want to display the original sentence)
    sentence = ''.join([dataset.vocab[idx] for idx in encoded_sentence.argmax(dim=1)])

    # Plotting the stroke
    f, ax = plt.subplots()

    x = np.cumsum(stroke[:, 1])
    y = np.cumsum(stroke[:, 2])

    size_x = x.max() - x.min() + 1.
    size_y = y.max() - y.min() + 1.

    f.set_size_inches(5. * size_x / size_y, 5.)

    cuts = np.where(stroke[:, 0] == 1)[0]
    start = 0

    for cut_value in cuts:
        ax.plot(x[start:cut_value], y[start:cut_value], 'k-', linewidth=3)
        ax.scatter(x[cut_value-1], y[cut_value-1], color='red', s=50)  # End of stroke marker
        start = cut_value + 1

    ax.axis('equal')
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)

    # Display the sentence below the plot
    plt.figtext(0.5, 0.01, sentence, wrap=True, horizontalalignment='center', fontsize=12)

    if save_name is None:
        plt.show()
    else:
        try:
            plt.savefig(save_name, bbox_inches='tight', pad_inches=0.5)
        except Exception as e:
            print("Error building image!: " + str(e))

    plt.close()
```
